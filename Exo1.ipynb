```python
#Le code suivant trace deux courbes, une représentant 
#la solution exacte et une représentant la solution approchée 
#par le schéma d'Euler explicite.


import numpy as np
import matplotlib.pyplot as plt 

"""Physical parameters"""

T = 10 # Temps final
y0 = 0.1 # Condition initiale
K = 10 # Donnée du problème


def f(t,y) :
    return y*np.log(K/y)

# Solution exacte
def y_e(t) :
    return K*(y0/K)**np.exp(-t)
    


"""Numerical parameters"""

N = 25 # Nombre de points de discrétisation


"""Numerical scheme"""

def explicit_euler(N,f,y0) :

    h = T/(N-1)  # Pas de temps
    t = np.linspace(0,T,N) # Discrétisation uniforme à N points du temps 
    #[0,T]: t[n] = t^n   si 0 <= n <= N-1
    y = np.zeros(N) # Création du vecteur où seront stockées les valeurs de 
    #la solution numérique: y[n] = y^n   si 0 <= n <= N-1
    y[0] = y0 # Initialisation avec condition initiale
    for n in range(0,N-1) :
        y[n+1] = y[n] + h*f(t[n],y[n]) # Formule Euler explicite pour calculer la solution au temps t^{n+1}
    return y


"""Post processing"""
ynum = explicit_euler(N,f,y0) # Résolution de l'EDO par Euler explicite

t = np.linspace(0,T,N) # Discrétisation uniforme du temps
plt.plot(t,ynum,label=r"$y_{num}$") # Affichage de la solution numérique
plt.plot(t,y_e(t),label=r"$y_{exact}$") # Affichage de la solution exacte
plt.xlabel(r"$t$") # Nom de l'axe x
plt.ylabel(r"$y(t)$") # Nom de l'axe y
plt.title("Explicit Euler") # Nom de la figure
plt.legend() # Affichage de la légende


#Les résultats obtenus :

# Les variables à changer en fonction de la question
#K = 10
#Tm = 10
#step = 0.1 # Alternera entre 0.1 et 1
#y0 = 0.1

# Les variables à changer en fonction de la question
#K = 10
#Tm = 10
#step = 0.1 # Alternera entre 0.1 et 1
#y0 = 2

#La limite egale a K sur les graphiques et on peut la vérifier par les calculs.
```

```python
#Question 2 : y'(t) =y²(t)

#On tracera à nouveau la solution exacte et la solution approchée par 
#schéma d'Euler explicite sur le même graphe.

import numpy as np
import matplotlib.pyplot as plt


w = 1
T = (2,0.49)[w-1] #T = 2
y0 = (0.2,2)[w-1] #y0 = 0.2

step = (0.1,0.001)[0] #step = 0.1


tn = [np.round(step*i,4) for i in range(int(T/step)+1)]

def exact_y(t):
    return 1/(1/y0 - t)

def f(t,y):
    return y**2


y = {0:y0}

for t in tn:
    y[np.round(t+step,4)] = (np.round(f(t,y[t])*step + y[t],4))
yn = list(y.values())
yn.pop(-1) # pour qu'elles ait la même dimension


t = np.linspace(0,T)
plt.plot(tn,yn, "r") # Affichage de la solution r numerique
plt.plot(t,exact_y(t), "b") # Affichage de la solution b exacte
plt.xlabel(r"$t$") # Nom de l'axe x
plt.ylabel(r"$y(t)$") # Nom de l'axe y

plt.title("Exo 1-2") # Nom de la figure

#on remarque que le premier graphe est incohérent.
#question = 1
#T = (2,0.49)[question-1]
#y0 = (0.2,2)[question-1]
#step = (0.1,0.001)[1] # (0.1,0.001)[1] pour l'image de droite

#question = 2
#T = (2,0.49)[question-1]
#y0 = (0.2,2)[question-1]

#step = (0.1,0.001)[1] # (0.1,0.001)[1] pour l'image de droite
```
