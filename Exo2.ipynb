```python
import numpy as np
import matplotlib.pyplot as plt

question = 1 # Toute cette section ne concerne que la question 1
y0 = (1,2)[1]
T = ((1.5,10)[0], 2*np.math.sqrt(y0))[question-1]
affichage_T = (T, f"2sqrt({y0})")[question-1]
step = 0.01


tn = [np.round(step*i,4) for i in range(int(T/step)+1)]


def exact_y(t):
    return ( 1/(1/y0 + t) , (np.math.sqrt(y0)-t/2)**2 )[question-1]

def f(t,y):
    return ( -y**2 , -np.math.sqrt(abs(y)) )[question-1]


y = {0:y0}

for t in tn:
    y[np.round(t+step,4)] = (np.round(f(t,y[t])*step + y[t],4))
yn = list(y.values())
yn.pop(-1) # pour qu'elles ait la même dimension


t = np.linspace(0,T) # Discrétisation uniforme du temps
plt.plot(tn,yn, "r") # Affichage de la solution numérique
plt.plot(t,exact_y(t), "b") # Affichage de la solution exacte
plt.xlabel(r"$t$") # Nom de l'axe x
plt.ylabel(r"$y(t)$") # Nom de l'axe y

plt.legend("Solution") # Affichage de la légende
```
